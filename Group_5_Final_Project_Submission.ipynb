```python
years = np.arange(2011, 2020)

plt.plot(years, US_f_survive, label = "Female")
plt.plot(years, US_m_survive, label = "Male")

plt.legend(bbox_to_anchor=(1.25, 1));

plt.yticks(np.arange(79, 90));
plt.xlabel('Year');
plt.ylabel('Percent Survival to Age 65 (%)');
plt.title('Percent Survival to Age 65 for Males and Females in the US');
plt.show()
```

It seems like females survive to age 65 at a higher percentage than males.

Question: Does percent survival to age 65 depend on gender in the US?

Null Hypothesis: Gender has no effect on percent survival to age 65.

Alternate Hypothesis: Being female has a positive effect on percent survival to age 65.

The statistic I am picking to assess the validity of the null hypothesis is the percent survival to age 65 for males and females in the United States from 2011 to 2019.

p = 0.05 for the significance level

```python
bootstrap_samples = 1000
US_m_means = []
US_f_means = []
difference_means = []

for i in np.arange(bootstrap_samples):
  US_m_sample_mean = np.mean(US_m_survive.sample(len(US_m_survive), replace = True))
  US_f_sample_mean = np.mean(US_f_survive.sample(len(US_f_survive), replace = True))

  US_m_means.append(US_m_sample_mean)
  US_f_means.append(US_f_sample_mean)

  difference_means.append(US_m_sample_mean - US_f_sample_mean)
```

```python
plt.title('Distribution of Bootstrapped Gender Differences in Survival to Age 65')
plt.hist(US_m_means, alpha = 0.5, label = 'male');
plt.hist(US_f_means, alpha = 0.5, label = 'female');
plt.legend(loc = 'upper center');
```

```python
np.percentile(US_m_means, 2.5), np.percentile(US_m_means, 97.5)
```

```python
np.percentile(US_f_means, 2.5), np.percentile(US_f_means, 97.5)
```

```python
plt.title('Distribution of Bootstrapped Diff in Means')
plt.hist(difference_means);
plt.ylabel('Frequency')
plt.xlabel('Male Survival to Age 65 - Female Survival to Age 65 (Difference)')
plt.axvline(x = 0, color = 'r')
plt.show()
```

```python
sum(np.array(difference_means) >= 0) / len(difference_means) #asks how often males survive longer than females (difference_means is negative, so the conditional will never be true)
```

The p value of my observed statistic is 0.0. I will therefore reject the null hypothesis in favor of the alternate hypothesis that being female has a positive effect on percent survival to age 65. There are many factors causing this, including: men tend to engage in more high-risk occupations (e.g. military, construction) and high-risk behaviors (e.g. drinking, gambling), women tend to make healthier lifestyle choices, and women tend to have strong social support networks.

## Hypothesis 4 (Josue)

Question: Is there a significant over-time change in the average male-female bachelors degree attainment proportions?

We define sex-differene to be: mean(male_attainment) - mean(female_attainment)

We define m to be our slope in the date-difference graph. This means m is the change of sex-difference bachelors degree attainment over time.

Null Hypothesis: There is no significant change over time of mean sex difference in bachelors attainment.

(m = 0)

```python
def standardize(anylist):
    '''convert any array of numbers to std units '''
    return (anylist - np.mean(anylist)) / np.std(anylist)
```

```python
def correlation(df, x, y):
    '''correlation between two columns of a dataframe '''
    return np.mean(standardize(df[x]) * standardize(df[y]))
```

```python
def slope(df, x, y):
    r = correlation(df, x, y)
    return r * np.std(df[y]) / np.std(df[x])
```

```python
def intercept(df, x, y):
    a = slope(df, x, y)
    return np.mean(df[y]) - a * np.mean(df[x])
```
